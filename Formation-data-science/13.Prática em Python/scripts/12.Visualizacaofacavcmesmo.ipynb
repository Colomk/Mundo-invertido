```python
# Importação das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
# Carregamento da base de dados
base = pd.read_csv('../dados/dados.csv', sep=';') # separador é ;
base.shape
```

```python
base.head() # visualização dos dados
```

```python
srn.histplot(base.PIB, kde=True, bins=10).set(title='PIB')
# Histograma com a densidade de probabilidade, com 10 divisões (bins) e título
# O histograma é uma representação gráfica de uma distribuição de frequências
# O eixo x representa a variável e o eixo y a frequência com que ela ocorre
# O kde é a linha que suaviza o histograma
```

```python
srn.histplot(base.VALOREMPENHO, kde=True, bins=10).set(title='Empenho')
```

```python
agrupado = base.sort_values('PIB').head(10) # ordena pelo PIB e pega os 10 primeiros
agrupado = agrupado.iloc[:,1:3] # pega somente as colunas 1 e 2
agrupado
agrupado.plot.bar(x='MUNICIPIO',y='PIB', color = 'gray') # gráfico de barras
```

```python
agrupado = base.sort_values('VALOREMPENHO').head(10)
agrupado = agrupado.iloc[:,[1,3]]
agrupado
agrupado.plot.bar(x='MUNICIPIO',y='VALOREMPENHO', color = 'gray')
```

```python
# Importação das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregamento da base de dados
base = pd.read_csv('../dados/trees.csv')
base.head()
```

```python
# Gráfico de dispersão considerando o volume e a dispersão
plt.scatter(x=base.Girth, y=base.Volume, color = 'blue', facecolors = 'none', marker = '*')
# Gráfico de linha considerando o volume e a dispersão
# x=base.Girth, y=base.Volume - são os dados que serão considerados
# color = 'blue' - cor dos pontos
# facecolors = 'none' - cor da face dos pontos
# marker = '*' - tipo de marcador

plt.title('Árvores') # título do gráfico
plt.xlabel('Circunferência') # nome do eixo x
plt.ylabel('Volume')    
```

```python
# Gráfico de linha considerando o volume e o atributo "girth" - sem marker
plt.plot(base.Girth,base.Volume) #marker removido
plt.title('Árvores')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
```

```python
# Gráfico de dispersão com 'afastamento' dos dados (jitter)
#fit_reg linha de tendência
sns.regplot(x=base.Girth,y=base.Volume, data = base, x_jitter = 0.3, fit_reg = False)
# X=base.Girth, y=base.Volume - são os dados que serão considerados
# data = base - fonte dos dados
# x_jitter = 0.3 - 'afastamento' dos dados
# fit_reg = False - não incluir a linha de tendência
```

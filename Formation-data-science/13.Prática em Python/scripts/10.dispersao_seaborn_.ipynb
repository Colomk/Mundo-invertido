```python
# Importação das bibliotecas
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
# Carregamento da base de dados
base = pd.read_csv('../dados/co2.csv')
base.head()
```

```python
# Gráfico de dispersão utilizando os atributos conc e uptake, agrupamento pelo type
srn.scatterplot(data=base, x='conc', y='uptake', hue='Type')
# data = base, x = coluna conc, y = coluna uptake, hue = agrupamento pelo type
```

```python
# Seleção de registros específicos da base de dados (Quebec e Mississipi)
q = base.loc[base['Type'] == 'Quebec'] # somente registros do Quebec
m = base.loc[base['Type'] == 'Mississippi'] # somente registros do Mississippi
```

```python
# Subgráfico (1 linha e duas colunas) mostrando gráficos sobre cada região
plt.figure() # cria uma nova figura
plt.subplot(1, 2, 1) # subplot com 1 linha e 2 colunas
srn.scatterplot(data=q, x='conc', y='uptake').set_title('Quebec')
# data = q, x = coluna conc, y = coluna uptake

plt.subplot(1, 2, 2) # subplot com 1 linha e 2 colunas
srn.scatterplot(data=m, x='conc', y='uptake').set_title('Mississippi')
# data = m, x = coluna conc, y = coluna uptake

plt.tight_layout() # ajusta o layout para não sobrepor os gráficos
plt.show() # mostra o gráfico
```

```python
# Refrigerado e não refrigerado
ch = base.loc[base['Treatment'] == 'chilled'] # Seleciona os registros da base de dados onde o tratamento é refrigerado
nc = base.loc[base['Treatment'] == 'nonchilled'] # Seleciona os registros da base de dados onde o tratamento é não refrigerado
```

```python
# Gráfico somente com 'chilled' e 'nonchilled'

plt.figure() # Cria uma nova figura
plt.subplot(1, 2, 1) # Subgráfico
srn.scatterplot(data=ch, x='conc', y='uptake').set_title('Chilled')
# data = ch, x = coluna conc, y = coluna uptake

plt.subplot(1, 2, 2)  # Subgráfico
srn.scatterplot(data=nc, x='conc', y='uptake').set_title('Non chilled')
# data = nc, x = coluna conc, y = coluna uptake

plt.tight_layout() # Ajusta o layout
plt.show() # Exibe o gráfico
```

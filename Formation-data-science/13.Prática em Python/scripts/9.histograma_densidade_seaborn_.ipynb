```python
# Importação das bibliotecas
import pandas as pd
import seaborn as srn
#pip install seaborn (executar no Anaconda Prompt)
import matplotlib.pyplot as plt
```

```python
# Carregamento da base de dados
base = pd.read_csv('../dados/trees.csv')
base.head()
```

```python
# Histograma com 10 divisões (bins) e com gráfico de densidade (kernel density - suaviação)
#stat: count, frequency, probability,percent, density
srn.histplot(base.Volume, color="blue", kde=True, stat="density", bins=10).set(title='Volume')
# Densidade
# color: cor do gráfico
# kde: suavização do gráfico, linha de densidade
# stat: tipo de densidade
# bins: divisões
# set(title='Título'): inclui o título no gráfico
```

```python
# Carregamento de outra base de dados
base2 = pd.read_csv('../dados/chicken.csv')
base2.head()
```

```python
# Criação de novo dataframe agrupando o atributo 'feed'
agrupado = base2.groupby(['feed'])['weight'].sum() # soma o peso por tipo de ração
agrupado
```
